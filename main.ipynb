## Model Selection

```python
from os import listdir, path

modelVersion = input('Select YOLO version model: ')
scale = input('Select the scale of the architecture: ')
archsPath = path.join('..', 'models', 'architectures')
weightQue = input('Do you to use pretrained wieghts? Y/N: ')
weightsPath = path.join('..', 'models', 'weights')

if modelVersion == '5':
    architecture = path.join(archsPath, 'yolov' + str(modelVersion) + scale + '.yaml')
    weights = path.join(weightsPath, 'yolov' + str(modelVersion) + scale + 'u.pt') if weightQue == 'Y' else None
else:
    architecture = path.join(archsPath, 'yolov' + str(modelVersion) + scale + '.yaml')
    weights = path.join(weightsPath, 'yolov' + str(modelVersion) + scale + '.pt') if weightQue == 'Y' else None

print("Selected Architecture is: ", architecture)
print("Selected Weights is: ", weights)
```

## Model Creation

```python
from ultralytics import YOLO
model = YOLO(architecture).load(weights) if not weights == None else YOLO(architecture)
```

## Model Training

```python
name = input('Insert a descriptive name of the model: ')
model.train(workers=1, device=0, data="../datasets/DsLMF_minerBehaviorDataset.yaml", epochs=200, lr0=0.01, lrf=0.01, momentum=0.937, batch=16, optimizer='SGD', plots =True, deterministic = False, project=path.join('models', 'trained', 'RTX3060'), name=name, close_mosaic=0, save_period=10, patience=20, verbose=False)
```

## Model Export

```python
from os import listdir, path
from ultralytics import YOLO

weightsPath = path.join('models', 'trained', 'v5n_Base', 'weights', 'best.pt')
print("Selected Weights is: ", weightsPath)

model = YOLO(weightsPath)
model.export(format='ncnn')
model.export(format='engine')
```
